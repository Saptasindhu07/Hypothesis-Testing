```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy.random import normal
import seaborn as sns
```

```python
sample1=normal(loc=0,scale=1,size=1000)
sample2=normal(loc=0,scale=1,size=1000)
sample3=normal(loc=0,scale=1,size=1000)
sample4=normal(loc=0,scale=1,size=1000)
sample5=normal(loc=0,scale=1,size=1000)
sample6=normal(loc=0,scale=1,size=1000)
```

```python
sns.kdeplot(sample1**2)
sns.kdeplot(sample1**2+sample2**2)
sns.kdeplot(sample1**2+sample2**2+sample3**2)
sns.kdeplot(sample1**2+sample2**2+sample3**2+sample4**2)
t=sample1**2+sample2**2+sample3**2+sample4**2+sample5**2
sns.kdeplot(t,clip=(t.min(),t.max()))
```

```python
def looper(degreeoffreedom):
    sum=0
    for i in range (degreeoffreedom):
        normaldistribution=normal(loc=0,scale=1,size=1000)
        sum+=normaldistribution**2
    sns.kdeplot(sum)
```

```python
looper(10)
```

```python
looper(50)
```

```python
looper(100)
```

```python
looper(1000)
```

```python
looper(100000)
```

**As we can see, distribution is closer to normal distribution as degree of freedom increases**
